```python
import pandas as pd

def parse_domtblout(file):
    # Lê o arquivo domtblout ignorando linhas de comentários
    data = []
    with open(file) as f:
        for line in f:
            if line.startswith("#"):
                continue
            parts = line.split()
            data.append({
                "Query": parts[0],
                "Target": parts[3],
                "E-value": float(parts[6]),
                "Score": float(parts[7]),
                "Domain": parts[8],
            })
    return pd.DataFrame(data)
```

```python
from scipy.stats import fisher_exact
import numpy as np

# Função para calcular as diferenças de prevalência e valores de p
def calculate_volcano_data(sample1, sample2):
    # Contagem de ocorrências por Target (Pfam)
    count_sample1 = sample1["Target"].value_counts()
    count_sample2 = sample2["Target"].value_counts()
    
    # Criar DataFrame unificado
    all_targets = set(count_sample1.index).union(count_sample2.index)
    data = pd.DataFrame(index=list(all_targets), columns=["Sample1", "Sample2"], data=0)
    data["Sample1"] = data.index.map(count_sample1).fillna(0)
    data["Sample2"] = data.index.map(count_sample2).fillna(0)
    
    # Total de sequências em cada amostra
    total_sample1 = len(sample1)
    total_sample2 = len(sample2)
    
    # Calcular diferenças de prevalência e valores de p
    prevalence_diff = []
    p_values = []
    for target in data.index:
        # Prevalência
        freq1 = data.loc[target, "Sample1"] / total_sample1
        freq2 = data.loc[target, "Sample2"] / total_sample2
        prevalence_diff.append(freq1 - freq2)
        
        # Tabela de contingência
        table = [
            [data.loc[target, "Sample1"], data.loc[target, "Sample2"]],
            [total_sample1 - data.loc[target, "Sample1"], total_sample2 - data.loc[target, "Sample2"]],
        ]
        _, p_value = fisher_exact(table)
        p_values.append(p_value)
    
    # Adicionar ao DataFrame
    data["Prevalence_Diff"] = prevalence_diff
    data["p_value"] = p_values
    data["-log10(p_value)"] = -np.log10(data["p_value"])
    return data
```

```python
volcano_data = calculate_volcano_data(sample1, sample2)
```
